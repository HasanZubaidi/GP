```python
#importing libraries for data integration
import pandas as pd
import os
import shutil
```

```python
#creating dataframes
dftr = pd.read_csv(r'C:\Users\SAMU\Desktop\Uni\GP\Bugbite Dataset\train\_annotations.csv')
dfts = pd.read_csv(r'C:\Users\SAMU\Desktop\Uni\GP\Bugbite Dataset\test\_annotations.csv')
dfv = pd.read_csv(r'C:\Users\SAMU\Desktop\Uni\GP\Bugbite Dataset\valid\_annotations.csv')

#defining the classes,images path, and destination
img_path = r'C:\Users\SAMU\Desktop\Uni\GP\Bugbite Dataset'
img_dest = r'C:\Users\SAMU\Desktop\Uni\GP\DermaTech\DermaTech Dataset'
```

```python
section = ['train','test','valid']

#moving training data
for index, row in dftr.iterrows():
    image_id = row['img_id']
    class_name = row['class']
    source = os.path.join(img_path,section[0],image_id)
    dest = os.path.join(img_dest,section[0],class_name)
    
    try:
        shutil.move(source, dest)
        print(f"Moved image {image_id} to {dest}")
    except FileNotFoundError:
        print(f"Image {image_id} not found in {source}")
    except Exception as e:
        print(f"Error moving image {image_id}: {e}")
```

```python
#moving test data
for index, row in dfts.iterrows():
    image_id = row['img_id']
    class_name = row['class']
    source = os.path.join(img_path,section[1],image_id)
    dest = os.path.join(img_dest,section[1],class_name)
    try:
        shutil.move(source, dest)
        print(f"Moved image {image_id} to {dest}")
    except FileNotFoundError:
        print(f"Image {image_id} not found in {source}")
    except Exception as e:
        print(f"Error moving image {image_id}: {e}")
```

```python
#moving validation data into test
for index, row in dfv.iterrows():
    image_id = row['img_id']
    class_name = row['class']
    source = os.path.join(img_path,section[2],image_id)
    dest = os.path.join(img_dest,section[1],class_name)
    try:
        shutil.move(source, dest)
        print(f"Moved image {image_id} to {dest}")
    except FileNotFoundError:
        print(f"Image {image_id} not found in {source}")
    except Exception as e:
        print(f"Error moving image {image_id}: {e}")
```
